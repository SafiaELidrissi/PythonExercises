```python
entries = []
#Openning the text file and skipping right to line 12
with open('ShowIpRoute.txt', 'r') as f:
    for i in range(11):
        f.__next__()
    for line in f:
        lst = f.readlines()
#Inserting the elements of each entry in the routing tables into a new list called line
for line in lst:
    entries.append(line.split())
#Displaying information about each entry depending on the condition performed upon the first member of the list line    
for line in entries:
    if (line[0] == 'O' and line[1] == 'E2'):
        prefix= line[2]
        AdMetric= line[3][1:-1]
        NextHop= line[5]
        LastUpdate= line[7][:-1]
        OutboundInt= line[8]
        print('Protocol: OSPF')
        print('Prefix: '+prefix)
        print('AD/Metric: '+AdMetric)
        print('Next Hop: '+NextHop)
        print('Last Update: '+LastUpdate)
        print('Outbound Interface: '+OutboundInt)
        print('\n')
        
    elif (line[0] == 'H'):
        prefix= line[1]
        AdMetric= line[2][1:-1]
        LastUpdate= line[5][:-1]
        OutboundInt= line[6]
        print('Protocol: HSRP')
        print('Prefix: '+prefix)
        print('AD/Metric: '+AdMetric)
        print('Next Hop: '+NextHop)
        print('Last Update: '+LastUpdate)
        print('Outbound Interface: '+OutboundInt)
        print('\n')
    elif (line[0] == 'C'):
        prefix= line[1]
        Mask= line[2]
        OutboundInt= line[6]
        print('Protocol: Directly connected route')
        print('Prefix: '+prefix)
        print('Mask: '+Mask)
        print('Outbound Interface: '+OutboundInt)
        print('\n')
    elif (line[0] == 'i'): 
        prefix= line[2]
        AdMetric= line[3][1:-1]
        NextHop= line[5]
        LastUpdate= line[7][:-1]
        OutboundInt= line[8]
        print('Protocol: IS-IS')
        print('Prefix: '+prefix)
        print('AD/Metric: '+AdMetric)
        print('Next Hop: '+NextHop)
        print('Last Update: '+LastUpdate)
        print('Outbound Interface: '+OutboundInt)
        print('\n')
    else:
        prefix= line[1]
        AdMetric= line[2][1:-1]
        NextHop= line[4]
        LastUpdate= line[6][:-1]
        OutboundInt= line[7]
        print('Protocol: EIGRP')
        print('Prefix: '+prefix)
        print('AD/Metric: '+AdMetric)
        print('Next Hop: '+NextHop)
        print('Last Update: '+LastUpdate)
        print('Outbound Interface: '+OutboundInt)
        print('\n')
#Closing the file        
f.close() 
```
